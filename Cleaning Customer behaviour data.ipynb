```python
#mapping mixed categories and item purchased
correct_mapping = {
    #clothing
    "T-shirt": "Clothing", "Shirt": "Clothing", "Shorts": "Clothing", "Hoodie": "Clothing", "Pants": "Clothing", "Socks": "Clothing", "Jeans": "Clothing", "Blouse": "Clothing", "Skirt": "Clothing", "Sweater": "Clothing", "Dress": "Clothing",

    #accessories
    "Sunglasses": 'Accessories', 'Gloves': 'Accessories', 'Jewelry': 'Accessories', 'Hat': 'Accessories', 'Handbag': 'Accessories', 'Backpack': 'Accessories', 'Belt': 'Accessories', 'Bag': 'Accessories', 'Scarf': 'Accessories',

    #electronics
    'Laptop': 'Electronics', 'Phone': 'Electronics', 'Watch': 'Electronics', 'Headphones': 'Electronics',

    #footwear
    'Shoes': 'Footwear', 'Sandals': 'Footwear', 'Sneakers': 'Footwear', 'Boots': 'Footwear',

    #outerwear
    'Coat': 'Outerwear', 'Jacket': 'Outerwear'
}
```

```python
df.head()
```

```python
# saving the changes via mapping
df['Category'] = df['Item Purchased'].map(correct_mapping)
```

```python
df.groupby('Category')['Item Purchased'].unique()
```

```python
df.isnull().sum()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#treating the null values of electronics since they come in standard size
df.loc[df['Category'] == 'Electronics', 'Size'] = 'Not Applicable'
```

```python
#as you can see it is already updated
df.groupby('Category')['Size'].unique()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#treating the null values of accessories
df.loc[df['Category'] == 'Accessories', 'Size'] = 'Free Size'
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#treating the null values of footwear
df.loc[df['Category'] == 'Footwear', 'Size'] = 'Not Applicable'
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#treating the null values of clothing
df.loc[(df['Category'] == 'Clothing') & (df['Size'].isnull()), 'Size'] = df[df['Category'] == 'Clothing']['Size'].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.head()
```

```python
#finding an average of each item to fill in null nalues for review ratings
df.groupby('Item Purchased')['Review Rating'].mean()
```

```python
#filling nulls in review rating based on the means of each item
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform ('median'))
```

```python
df.isnull().sum()
```

```python
df['Review Rating']= df['Review Rating'].round(2)
```

```python
df.head()
```

```python
#handling nulls in purchased amount usd
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform ('median'))
```

```python
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].round(2)
```

```python
df.isnull().sum()
```

```python
#handling the missing values for previous purchases
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

## Renaming the header so to import the data to SQL using underscores

```python
#Renaming the columns so that the data can be imported to SQL, and making them lower case
df.columns= df. columns.str.replace(' ', '_')
df.columns= df. columns.str.lower()
```

```python
#renaming the purchase amount column since it is long
df=df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```
